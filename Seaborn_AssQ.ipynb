Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Here are five popular plots that can be created using Seaborn and their uses:

Histograms: A histogram is used to visualize the distribution of a continuous variable. It is a bar-like graph where the x-axis represents the range of values and the y-axis represents the frequency or count of observations within each range. Seaborn's distplot() function can create histograms.

Scatter Plots: A scatter plot is used to visualize the relationship between two continuous variables. It is a plot where each data point is represented by a point on the graph with the x-coordinate and y-coordinate representing the two variables. Seaborn's scatterplot() function can create scatter plots.

Bar Plots: A bar plot is used to compare the values of a categorical variable using rectangular bars where the height of each bar represents the value of the variable. Seaborn's barplot() function can create bar plots.

Line Plots: A line plot is used to visualize the relationship between two continuous variables, similar to a scatter plot, but with the points connected by lines. It is useful for showing trends over time or across other continuous variables. Seaborn's lineplot() function can create line plots.

Heatmaps: A heatmap is used to visualize the relationship between two categorical variables or one categorical and one continuous variable. It is a plot where the color of each cell represents the value of the variable at the intersection of its row and column. Seaborn's heatmap() function can create heatmaps.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
sns.distplot(tips['total_bill']) #Histogram
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips) #Scatter plot
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=tips) #Bar plot
plt.show()
```

```python
sns.lineplot(x='day', y='total_bill', data=tips) #line plot
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True) # Heatmap
plt.show()
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create a line plot using the 'lineplot' function
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

# Add a title and axis labels
plt.title("FMRI Signal over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Display the plot
plt.show()
```

In this code, we first load the "fmri" dataset using the load_dataset() function of Seaborn. We then create a line plot using the lineplot() function, specifying the x-axis as "timepoint", the y-axis as "signal", and using the "event" and "region" columns to differentiate the lines for each event and region. We then add a title and axis labels using plt.title(), plt.xlabel(), and plt.ylabel(), respectively. Finally, we display the plot using plt.show().

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 6))

# Box plot for 'age' by 'pclass'
sns.boxplot(x="pclass", y="age", data=titanic, ax=ax1)
ax1.set_title("Age by Passenger Class")

# Box plot for 'fare' by 'pclass'
sns.boxplot(x="pclass", y="fare", data=titanic, ax=ax2)
ax2.set_title("Fare by Passenger Class")

# Display the plot
plt.show()
```

In this code, we first load the "titanic" dataset using the load_dataset() function of Seaborn. We then create a figure with two subplots using plt.subplots(), specifying the number of columns as 2 and the figure size as (12, 6). We then create a box plot for 'age' and 'fare' using the boxplot() function, specifying the x-axis as 'pclass' and the y-axis as the respective column, and passing in the 'titanic' dataset. We also set the title for each subplot using ax.set_title(). Finally, we display the plot using plt.show().

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram of 'price' with 'cut' as a hue
sns.histplot(data=diamonds, x="price", hue="cut")

# Add a title and axis labels
plt.title("Price Distribution by Cut")
plt.xlabel("Price")
plt.ylabel("Count")

# Display the plot
plt.show()
```

In this code, we first load the "diamonds" dataset using the load_dataset() function of Seaborn. We then create a histogram of the 'price' column using the histplot() function, specifying the data as 'diamonds', the x-axis as 'price', and the hue as 'cut'. We then add a title and axis labels using plt.title(), plt.xlabel(), and plt.ylabel(), respectively. Finally, we display the plot using plt.show().

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a pair plot of all columns with 'species' as a hue
sns.pairplot(data=iris, hue="species")

# Display the plot
plt.show()
```

In this code, we first load the "iris" dataset using the load_dataset() function of Seaborn. We then create a pair plot of all columns using the pairplot() function, specifying the data as 'iris' and the hue as 'species'. This creates a grid of scatter plots for each pair of columns, with the points colored according to the species of iris. Finally, we display the plot using plt.show().

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.


```python
import seaborn as sns

# Load the flights dataset
flights = sns.load_dataset("flights")

# Reshape the dataset into a pivot table
flights = flights.pivot("month", "year", "passengers")

# Create a heatmap of the pivot table
sns.heatmap(flights, cmap="YlGnBu")

# Add a title and axis labels
plt.title("Passenger Traffic by Year and Month")
plt.xlabel("Year")
plt.ylabel("Month")

# Display the plot
plt.show()
```

In this code, we first load the "flights" dataset using the load_dataset() function of Seaborn. We then reshape the dataset into a pivot table using the pivot() function, specifying the row index as 'month', the column index as 'year', and the values as 'passengers'. We then create a heatmap of the pivot table using the heatmap() function, specifying the color map as 'YlGnBu'. We also add a title and axis labels using plt.title(), plt.xlabel(), and plt.ylabel(), respectively. Finally, we display the plot using plt.show(). The resulting heatmap shows the passenger traffic by year and month, with darker colors indicating higher passenger counts.
